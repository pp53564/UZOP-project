
U DBSCAN (Density-Based Spatial Clustering of Applications with Noise), točke koje pripadaju istom klasteru nazivaju se "povezanima" ili "dosljednima". DBSCAN klasificira točke na temelju gustoće, što znači da točke s dovoljno "susjeda" u blizini čine klaster.
Core points (jezgra): To su točke koje imaju najmanje min_samples susjeda unutar radijusa eps. Core points često tvore središte klastera.

Border points (rubne točke): To su točke koje se ne kvalificiraju kao jezgre jer nemaju dovoljno susjeda za definiranje klastera, ali su unutar radijusa eps neke jezgre. Rubne točke često pripadaju klasteru, ali nisu središnje točke klastera.

Noise points (šum): To su točke koje nisu ni jezgre ni rubne, tj. nemaju dovoljno susjeda niti su same susjed nekoj jezgri. Šumne točke obično nisu dio klastera. 

```python
#DBSCAN
df = pd.read_csv('estonia-passenger-list-modified.csv') 
df.loc[df.Age.isna(), 'Age'] = df.Age.median()

X = df[['Age']].values
dbscan = DBSCAN(eps=4, min_samples=2)
clusters = dbscan.fit_predict(X)
df['cluster'] = clusters # rjesenje

plt.scatter(df.index, df['Age'], c=df['cluster'])
plt.xlabel('Index')
plt.ylabel('Vrijednost')
plt.title('DBSCAN na jednom stupcu')
plt.show()

print(df.loc[df.cluster == -1, 'Age'])
```

```python
import os

import torchvision
from torchvision.io import read_image

from torch.utils.data import Dataset

class EuroSATDataset(Dataset):
  def __init__(self, data_dir):
    self.instances = []

    curr_label = -1
    for subdir, _, files in os.walk(data_dir):
      if curr_label < 0:
        curr_label += 1
        continue

      for file in files:
        if file.endswith('.jpg'):
          self.instances.append((os.path.join(subdir, file), curr_label))
      curr_label += 1
          
  def __len__(self):
    return len(self.instances)

  def __getitem__(self, idx):
    image = read_image(self.instances[idx][0]) / 255.0
    label = self.instances[idx][1]-1
    
    return image, label
```

```python


class MLPModel(nn.Module):
    def __init__(self, img_dims=(28, 28), num_classes=10):
        super(MLPModel, self).__init__()
        self.linear1 = nn.Linear(img_dims[0] * img_dims[1], 10)
        self.linear2 = nn.Linear(10, 1)

    def forward(self, x):
        x = F.relu(self.linear1(x))
        x = self.linear2(x)

        return x
```

```python
class ConvModel(nn.Module):
  def __init__(self, img_dims=(28, 28), pool_kernel_size=3, num_classes=10):
    super(ConvModel, self).__init__()

    self.pool_kernel_size = pool_kernel_size

    self.conv1 = nn.Conv2d(1, 24, 5, 1, padding='same')
    self.conv2 = nn.Conv2d(24, 48, 5, 1, padding='same')

    self.linear1 = nn.Linear(48 * (img_dims[0]//pool_kernel_size**2) * (img_dims[1]//pool_kernel_size**2), 10)

  def forward(self, x):
    x = F.relu(self.conv1(x))
    x = F.max_pool2d(x, self.pool_kernel_size, self.pool_kernel_size)
    x = F.relu(self.conv2(x))
    x = F.max_pool2d(x, self.pool_kernel_size, self.pool_kernel_size)

    x = torch.flatten(x, start_dim=-3)

    x = self.linear1(x)

    return x
```

```python
def eval(model, dataloader):
  model.eval()
  with torch.no_grad():
    eval_loss = []
    correct = 0
    count = 0

    for eval_batch in dataloader:
      x, y = eval_batch
      #
      #x = x.cuda(0)
      #y = y.cuda(0)
      logits = model(x)
      loss = F.cross_entropy(logits, y)
      eval_loss.append(loss)
      correct += (logits.argmax(dim=-1) == y).float().sum()
      count += len(y)

    eval_loss = torch.mean(torch.tensor(eval_loss))
    eval_acc = correct / count
  return eval_loss.item(), eval_acc.item()

def train(model, train_dataloader, val_dataloader, test_dataloader, num_epochs=1, lr=1e-3):
  #
  #model.cuda(0)
  optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=0.1)

  # TRAINING
  model.train()
  val_losses = []
  train_losses = []

  val_loss, val_acc = eval(model, val_dataloader)
  val_losses.append(val_loss)

  for epoch in range(num_epochs):
    with tqdm(enumerate(train_dataloader), total=len(train_dataloader), desc=f'Training (epoch={epoch+1}/{num_epochs})') as epoch_progress:
      for batch_idx, train_batch in epoch_progress:
        x, y = train_batch
        #
        #x = x.cuda(0)
        #y = y.cuda(0)
        logits = model(x)
        loss = F.cross_entropy(logits, y)
        train_losses.append(loss.item())
        epoch_progress.set_postfix({'train loss': loss.item()})

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    val_loss, val_acc = eval(model, val_dataloader)
    val_losses.append(val_loss)
    print(f"Validation (epoch={epoch+1}/{num_epochs}): loss={val_loss}, acc={val_acc}")
```
